### Model Evaluation

```python
# import sk learn train test split
from sklearn.model_selection import train_test_split

# Assign variables to capture train test split output
X_train, X_test, y_train, y_test = train_test_split(X,Y)
```

```python
#Import SK learn linear regression
from sklearn.linear_model import LinearRegression

# Instantiate
linreg = LinearRegression()

# Fit model to training data 
linreg.fit(X_train, y_train) 
```

```python
#Assign feature cols
feature_cols = ['Days_Open','News_Count','Avg_Tmp','WeekDay','big_data_newsCount']

#print y-intercept
print(linreg.intercept_)

#print the beta coefficients in same order as passed 
print(linreg.coef_)

# Zip can pair feature names and coefficients together
zip(feature_cols, linreg.coef_)
```

```python
y_pred = linreg.predict(X_test)
```

```python
import numpy as np

#import metrics library
from sklearn import metrics

#print results of MAE
print(metrics.mean_absolute_error(y_test, y_pred))

# Print results of MSE
print(metrics.mean_squared_error(y_test, y_pred))

# Print result of RMSE
print(np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
