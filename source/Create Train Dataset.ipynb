```python
import pandas as pd
import numpy as np 
```

```python
train = pd.read_csv("../Data/train2.csv", encoding = "EUC-KR")
```

```python
train.head()
```

```python
train.columns = ["Year", "Count", "Days_Open", "Avg_Count", "News_Count", "Start_date", "End_date", "Profit", "Expense", "Temp", "Vis_change_rate", "big_data_newsCount","prev_count","mid_count","after_count"]
```

```python
train.head()
```

```python
g = [x[0:4] for x in train.Year]
train.Year = g
train.head()
```

```python
train.index = train["Start_date"]
```

```python
train.drop("Start_date", axis = 1)
```

```python
train.head()
```

```python
g = [x[5:7] for x in train.Start_date]
train["Month"] = g
train.head()
```

```python
g = [x[8:10] for x in train.Start_date]
train["Day"] = g
train.head()
```

```python
import datetime
train["WeekDay"] = (pd.to_datetime(train.Start_date,format='%Y-%m-%d')).dt.weekday
train.head()
```

```python
train.groupby(["WeekDay"]).mean()
```

```python
train.groupby(["Month"]).mean()
```

```python
train.groupby(["Days_Open"]).mean()
```

```python
train.head()
```

```python
train.drop("Profit", axis = 1, inplace=True)
train.drop("Expense", axis = 1, inplace=True)
train.drop("Temp", axis = 1, inplace=True)
```
